Empezaremos viendo el IDLE. Abrimos la aplicacin *IDLE Python 3.8* y nos aparecer la misma ventana que vieron arriba. El IDLE es como una calculadora en vivo que puede realizar operaciones y devuelve el resultado al instante. Por ejemplo, si pongo:

```python
50 - 5*6
```

Apretando enter nos sale el resultado y nos aparecen tres signos **<font color = "red"> >>> </font>** de color granate. Esto indica que nuestro clculo ha sido procesado automticamente. Veamos otro ejemplo:

```python
(50 - 5*6)/4
```

El signo igual (=) se usa para asignar un valor a una variable. Las asignaciones de variables no muestran ningn resultado al usuario.

```python
alto = 20
ancho = 5
alto * ancho
```

Observemos varias cosas importantes:

1. Yo elijo el nombre de las variables
2. Python memoriza el valor de las variables que se definen anteriormente, hasta que cerremos el IDLE de nuevo.
3. Si preguntamos por una variable que no ha sido asignada an, obtendremos un error

Veamos este ltimo punto en accin. Cuando realizamos algo que Python no entiende, nos informar de un error:

```python
n
```

Hay varios componentes presentes en un error, pero sin dudas a veces es mucho ms simple entender el error que corregirlo. __Slo la prctica nos ayudar a evitar los errores y a corregirlos ms rapidamente__. 

Cuando recibamos un error, lo primero que debemos hacer es, paradjicamente, observar la ltima lnea:
 > `NameError: name 'n' is not defined`

Lastimosamente al da de hoy la inmensa mayora de los errores de Python estn en ingls. Es mejor que nos vayamos acostumbrando a eso. Este error nos dice lo que nos esperbamos: La variable `n` no se ha definido antes, y por lo tanto no tiene un valor asignado, y Python no entiende que hacer con esa variable.  En general, con mirar la ltima lnea del error ya debera bastar. Veamos otro error comn:

```python
var = 2
x = 0
y = var / x
```

Asignamos los valores $2$ y $0$ a las variables $var$ y $x$. Queremos asignar a la variable $y$ el valor de la divisin entre $var$ y $x$, pero como $x$ es cero, entonces $\frac{var}{x}$ es $\frac{2}{0}$. Sin embargo, no est permitido dividir por cero (No tiene sentido!), y eso es lo que nos dice la ltima lnea, justamente. 

Parece broma, pero *la forma ms natural de conocer todos los errores es cometindolos uno mismo*. Es parte del proceso de aprender.

### El Script de Python

Todo lo que hemos hecho hasta ahora ha sido usando el *Python IDLE* (alias "La calculadora"). Ahora veamos el otro interfaz, que es el *Script*. Accedemos a un script apretando `Ctrl + N` cuando estemos en el *Python IDLE*. Nos abrir un archivo completamente nuevo y vaco. El *Script*, a diferencia del *IDLE*, se guarda en el disco duro de nuestra computadora y podemos accederlo y editarlo inclusive luego de cerrar Python. Sin embargo, no es interactivo, si queremos que python analice o lea lo que escribimos en el Script, debemos apretar `F5`. Igual que un archivo Excel tiene la extensin *.xlsx*, las imgenes son generalmente *.jpg* o *.png*, los Scripts de Python tienen la extensin *.py*. stos no son ms que un conjunto de instrucciones que Python est listo para leer cuando apretemos `F5`.

Otra funcionalidad que es interesante en los scripts son los comentarios. Estas son lineas de codigo que no son ledas por Python, son completamente ignoradas. Estn hechas para que lo lean los humanos y el cdigo sea ms legible. Cualquier lnea que empiece con un numeral `#` es un comentario, y podemos escribir lo que queramos:

```python
#Esto es un comentario, no necesita tener sentido, ser ignorado por Python. 
```

Si queremos que el comentario ocupe ms de una lnea, debemos encerrar el texto por triples comillas simples (''')

```python
''' Esto es un 
comentario

multilinea. Puede ser tan largo como yo quisiera.

'''
```

Los resultados que obtengamos en un script se mostrarn en el IDLE, siempre. La siguiente tabla encapsula las diferencias entre ambos:

| Script | IDLE |
| ----------- | ----------- |
| Guarda las variables  | Olvida las variables |
|Instrucciones por bloques | Instrucciones por linea|
| Hay que apretar `F5` | Hay que apretar `Enter` |
